```python
# Import necessary library

# For managing audio file
import librosa

```

```python

#Importing Pytorch
import torch
```

```python

#Importing Wav2Vec
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
# Reading taken audio clip

import IPython.display as display
#display.Audio("taken_clip.wav", autoplay=True)
display.Audio("1artificial.wav", autoplay=True)
```

```python
# Loading the audio file

audio, rate = librosa.load("1artificial.wav", sr = 16000)
```

```python
# printing audio 
print(audio)

```

```python
# printing rate
print(rate)
```

```python
# Importing Wav2Vec pretrained model

tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
# Taking an input value

input_values = tokenizer(audio, return_tensors = "pt").input_values
```

```python
input_values
```

```python
# Storing logits (non-normalized prediction values)
logits = model(input_values).logits
```

```python
# Storing predicted ids
prediction = torch.argmax(logits, dim = -1)
```

```python
prediction
```

```python
type(prediction)
```

```python
len (prediction[0])
```

```python
# Passing the prediction to the tokenzer decode to get the transcription
transcription = tokenizer.batch_decode(prediction)[0]
```

```python
# Printing the transcription
print(transcription)
```
