```python
from models.i3d.extract_i3d import ExtractI3D
from utils.utils import build_cfg_path, action_on_extraction
from omegaconf import OmegaConf
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.cuda.get_device_name(0)
```

```python
train_df = pd.read_csv("video.csv")
root_dir = '06/'
train_df.filename = root_dir + train_df.filename
videos = list(train_df.filename[9000:])
```

```python
# Select the feature type
feature_type = 'clip'
model_name = 'ViT-B/32'

# Load and patch the config

args = OmegaConf.load(build_cfg_path(feature_type))
args.feature_type = feature_type
#args.video_paths = ['./sample/v_GGSY1Qvo990.mp4']
#args.video_paths = ['../../1.mp4', '2.mp4', '3.mp4']
args.video_paths = videos.copy()
# args.show_pred = True
# args.stack_size = 24
# args.step_size = 24
# args.extraction_fps = 25
args.flow_type = 'raft' # 'pwc' is not supported on Google Colab (cupy version mismatch)
# args.streams = 'flow'

# Load the model
extractor = ExtractCLIP(args)
modules_dict = extractor.load_model(device)
print(modules_dict.keys())
```

```python
# Extract features
list_video_features=[]


for video_path in args.video_paths:
    print(f'Extracting for {video_path}')   
    features = extractor.extract(device, modules_dict, video_path)
    #[(print(k), print(v.shape), print(v)) for k, v in features.items()]
    #action_on_extraction(features, video_path, output_path=root_dir+'output', on_extraction='save_numpy')
    #[(print(k), print(v)) for k, v in features.items()]
    #print(features['rgb'][0])
    #print(features['flow'][0])
    #print (len(features['rgb'][0]))
    #print (len(features['flow'][0]))
    #[(print(k), print(v.shape), print(v)) for k, v in features.items()]
    a = list(features['clip'][0])
    #print (a)
    list_video_features.append(a)
    #list_video_features.append(features['rgb'],features['rgb'])

list_video_features = pd.DataFrame(list_video_features)
#save(root_dir+'video_i3d_features_0-10.npy', list_video_features)



```
